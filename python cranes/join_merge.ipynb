```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['s1','s1','s2','s2','s1','s2','s2','s1'],
    'Sales':[100,200,150,250,120,180,200,300],
    'Quantity':[10,15,12,18,8,20,15,25],
    'Date':pd.date_range('2023-01-01',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat_sales_sum = df.groupby('Category')['Sales'].sum()
cat_sales_sum
```

```python
sto_sales_sum = df.groupby('Store')['Sales'].sum()
sto_sales_sum
```

```python
multi_sales_sum = df.groupby(['Category','Store'])['Sales'].sum()
multi_sales_sum
```

```python
multi_sales_df = multi_sales_sum.reset_index()
multi_sales_df
```

```python
df['Sales'].agg(['mean','median','min','max','std','count','var'])
```

```python
grouped_agg = df.groupby('Category').agg({
    'Sales':['sum','max'],
    'Quantity':['mean','median']
})

grouped_agg
```

```python
import pandas as pd

employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

```python
data ={
    'Date':pd.date_range('2023-01-01',periods=20),
    'Product':['A','B','C','D']*5,
    'Region':['East','West','North','South','East','West','North','South','East','West','North','South','East','West','North','South','East','West','North','South'],
    'Sales':np.random.randint(100,1000,20),
    'Units':np.random.randint(10,100,20),
    'Rep':['John','Marry','Bob','Alice','John','Marry','Bob','Alice','John','Marry','Bob','Alice','John','Marry','Bob','Alice','John','Marry','Bob','Alice']
}
df = pd.DataFrame(data)
df
```

```python
df['Month']=df['Date'].dt.month_name()
df['Quarter']='Q'+df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median')
```

```python
pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
```

```python
pd.pivot_table(df,values='Sales',index=['Quarter','Region'],columns='Product',aggfunc='sum')
```

```python
pd.crosstab(df['Region'],df['Product'])
```

```python
pd.crosstab(df['Region'],df['Product'],normalize='index')
```
