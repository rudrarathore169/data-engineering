## Data Exploration and Data Cleaning

#### Missing Values

##### Missing values are identified to determine which columns require cleaning before analysis

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ","_")
      .str.replace(r"[()]","",regex=True)
)
```

```python
df.columns
```

```python
df.isnull().sum()
```

### Check Null Values Percentage

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
```

```python
missing_percentage.sort_values(ascending=False)
```

#### Observation - This tells you which columns have the most missing values.

### Handling Missing Values Numerical

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

### Observation - Filtering out the Numerical columns and Categorical Columns

### Fill Missing Values in Categorical Values

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

#### Observation All numeric columns now have no missing values — every NaN has been replaced with the median of its respective column

## Outlier Detection and Handling


### We will use the Z-score method to identify outliers in key numerical variables and visualize them using Box Plots.

```python
import numpy as np
selected_cols = ['age', 'salary_usd', 'experience_years']


plt.figure(figsize=(15, 5))
for i, col in enumerate(selected_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
z_scores = (df[selected_cols] - df[selected_cols].mean()) / df[selected_cols].std()
outliers_mask = (np.abs(z_scores) > 3).any(axis=1)

print(f"Total rows with outliers (Z-score > 3): {outliers_mask.sum()}")

if outliers_mask.any():
    display(df[outliers_mask].head())
```

#### *   Detection Methodology: Used Box Plots for visual identification and the Z-Score method (threshold > 3) to statistically flag 256 rows as significant outliers in age, salary, and experience.

#### *   Validation: Confirmed that these outliers represent legitimate extreme values in the dataset rather than data entry errors, ensuring the integrity of the subsequent analysis.

## Exploratory Data Analysis (EDA)

#### In this section, we solve 15 problem statements to uncover insights into AI adoption and productivity.

### Question 1: Which industries achieve the highest average productivity improvement after adopting AI?

```python
mapping = {'Increased': 1, 'Same': 0, 'Decreased': -1}
df['productivity_numeric'] = df['productivity_change'].map(mapping)

industry_productivity = df.groupby('industry')['productivity_numeric'].mean().sort_values(ascending=False)

industry_productivity
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=industry_productivity.values,
    y=industry_productivity.index,
    hue=industry_productivity.index,
    palette='Set2',
    legend=False
)

plt.title("Average Productivity Improvement by Industry")
plt.xlabel("Average Productivity Change")
plt.ylabel("Industry")

plt.show()
```

#### Observation - The chart highlights the industries that experience the highest productivity improvement after adopting AI tools. Industries with higher average productivity gains may be leveraging AI more effectively in their daily operations.

### Question 2: Which AI tool saves the maximum average time per day for employees?

```python
tool_usage = df["primary_ai_tool"].value_counts()

tool_usage
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    tool_usage,
    labels=tool_usage.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribution of AI Tool Adoption")

plt.show()
```

#### Observation -The pie chart illustrates the proportion of employees using each AI tool, helping identify the most widely adopted tools in the dataset..
