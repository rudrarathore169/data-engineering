```python
import numpy as np
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
```

```python
z_scores = np.abs((df['tip']-df['tip'].mean())/df['tip'].std())
print(z_scores)

# Outliers where z>3

outliers_zscore = df[z_scores>3]
print("Outliers detected using Z-score:")
print(outliers_zscore)

# Scatter plot tups with outliers marked

plt.figure(figsize=(8,5))
plt.scatter(range(len(df)),df['tip'],label='data',alpha=0.7)

#HIghlight outliers

plt.scatter(outliers_zscore.index,outliers_zscore['tip'],color='red',label='Z-score Outliers')
plt.title("scatter plot of tups with z-score Outliers")
plt.xlabel("Index")
plt.ylabel("Tip Amount")
```

```python
#remove outliers using Z-score
#Keep only rows where z-score<3

df_no_outliers_zscore=df[z_scores<3]
print("data size after removing z-scores outliers:",len(df_no_outliers_zscore))
```

```python
# z-score calculation
z_scores = np.abs((df['tip']-df['tip'].mean())/df['tip'].std())
# remove outliers
df_no_outliers_zscore=df[z_scores<3]
# scatter plot of cleaned data
plt.figure(figsize=(8,5))
plt.scatter(range(len(df_no_outliers_zscore)),df_no_outliers_zscore['tip'],
                   label="Cleaned Data",alpha=0.7,color="blue")
plt.title("scatter Plot of tips (after z-score Outlier Removal)")
plt.xlabel("Index")
plt.ylabel("Tip Amount")
plt.legend()
plt.show()
```
