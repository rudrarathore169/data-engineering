```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data_clean.csv')
```

```python
data['Ozone']  = data['Ozone'].fillna(data['Ozone'].mean())
```

```python
data.isnull().sum()
```

```python
data['Solar.R']  = data['Solar.R'].fillna(data['Solar.R'].mean())
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data = data.drop('Temp C', axis=1)
```

```python
data.head()
```

```python
data.isnull().sum().sum()
```

```python
data['Weather'] = data['Weather'].ffill()
```

```python
data.isnull().sum()
```

```python
data['Month'] = data['Month'].replace('May', 5).astype(int)

data['date'] = pd.to_datetime({
    'year': data['Year'],
    'month': data['Month'],
    'day': data['Day']
})
```

```python
data.head()
```

```python
data = data.drop(['Day', 'Month', 'Year'], axis=1)
```

```python
data.head()
```

```python
hottest_days = data.sort_values(by='Temp',ascending=False).head(5)
print(hottest_days)
```

```python
avg_ozone = data.groupby('Weather')['Ozone'].mean().sort_values(ascending=False)
print(avg_ozone)
```

```python
temp_avg = data[
    (data['Temp'] > data['Temp'].mean()) &
    (data['Ozone'] < data['Ozone'].mean())
]

print(temp_avg)
```

```python
data['Heat Level'] = np.where(data['Temp'] > 80, 'High', 'Low')
```

```python
data.head()
```

```python
avg_solar = data.groupby(data['date'].dt.month_name())['Solar.R'].mean()

print(avg_solar)
print("Month with maximum average Solar Radiation:", avg_solar.idxmax())
print("Average Solar Radiation:", avg_solar.max())
```
