```python
x= np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x,y,label="sin(x)")
plt.title("Simple Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
fig,ax  = plt.subplots(2,2)
title=["Linear","Double","Square","Cube"]
y = [x,x*2,x*x,x*x*x]

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(x,y[i])

fig.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(7,3))
x1 = [1,2,3,4,5,6]
y1=[45,34,30,45,50,38]
y2=[36,28,30,40,38,48]

labels=["student 1","student 2"]

fig.suptitle('Student marks in different subjects', fontsize=10)
l1 = ax1.plot(x1,y1,'o-',color='g')
l2 = ax2.plot(x1,y2,'o-')
fig.legend([l1,l2],labels=labels,loc="upper right")
plt.show()
```

```python
day = [1,2,3,4,5,6,7,8,9]
rating = [3.5,4,2,5.5,8,7.5,10,9,9.5]
n= np.array(['r','b','g','y'])
ind = np.random.randint(0,len(n),len(day))
plt.scatter(day,rating,color=n[ind],s=500,alpha=0.7,marker='^',edgecolor='m',linewidth=2)
plt.xlabel('DAY',fontsize=15)
plt.ylabel('RATING',fontsize=15)
plt.title('RESTAURANT',fontsize=30)
plt.show()
```

```python
x= np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000*np.random.rand(100)

plt.figure(figsize=(10,6))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5)
plt.title("Scatter Plot With Random Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.colorbar(label="Color Values")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(batsman["index"],batsman["RG Sharma"],color = 'blue',marker='o',label='RG Sharma')
plt.scatter(batsman["index"],batsman["V Kohli"],color = 'red',marker='^',label='V Kohli')

plt.title("Scatter Plot: Sharma vs Kohli Runs",fontsize=14,fontweight='bold')
plt.xlabel('years')
plt.ylabel('Runs')
plt.legend()
plt.grid(True,linestyle='--',alpha=0.6)
```

```python
plt.figure(figsize=(10,6))
bar_width=0.4
plt.bar(batsman["index"]-0.2,batsman["RG Sharma"],width=bar_width,color="skyblue",label = "RG Sharma")
plt.bar(batsman["index"]+0.2,batsman["V Kohli"],width=bar_width,color="salmon",label = "V Kohli")

plt.title("Bar plot: yearly Runs Comparison",fontsize=14,fontweight='bold')
plt.xlabel('year')
plt.ylabel('runs')
plt.xticks(batsman['index'])
plt.legend()
plt.grid(axis="y",linestyle='--',alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10,6))
bar_width=0.4
plt.bar(batsman["index"]-0.2,batsman["RG Sharma"],width=bar_width,color="skyblue",label = "RG Sharma")
plt.bar(batsman["index"]+0.2,batsman["V Kohli"],bottom=batsman["RG Sharma"],color="salmon",label = "V Kohli")

plt.title("Bar plot: yearly Runs Comparison",fontsize=14,fontweight='bold')
plt.xlabel('year')
plt.ylabel('runs')
plt.xticks(batsman['index'])
plt.legend()
plt.grid(axis="y",linestyle='--',alpha=0.6)
plt.show()
```

```python
total_runs=[batsman["RG Sharma"].sum(),batsman["V Kohli"].sum()]
players = ["RG Sharma","V Kohli"]
colors = ["lightblue","lightcoral"]

plt.figure(figsize=(6,6))
plt.pie(total_runs,labels=players,autopct="%1.1f%%",startangle=90,colors=colors,explode=(0.05,0.05))
plt.title("Pie Chart: Overall Contribution (2008-2017)",fontsize=14,fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(batsman["RG Sharma"],bins=5,alpha=0.7,color='blue',label="RG Sharma")
plt.hist(batsman["V Kohli"],bins=5,alpha=0.7,color='Red',label="V Kohli")
plt.title("HISTOGRAM: Distribution of Runs", fontsize=14,fontweight="bold")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis="y",linestyle='--',alpha=0.6)
plt.show()
```
